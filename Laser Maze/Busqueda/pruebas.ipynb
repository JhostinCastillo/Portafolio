```python
# La funcion objetivo y la funcion de expansion reciben un nodo de entrada

# Funcion objetivo. En este caso seria la funcion que verifica todas las metas y checkpoints
goal = lambda x: x.dato=='ABZB' # Devuelve un booleano

# Funcion de expansion. Sera una funcion que seleccione algun movimiento en el juego
def F(nodo):            # Devuelve una lista de nodos
    vecinos = []
    valor = nodo.dato
    # Generamos dos nuevos nodos
    for nuevo_valor in [valor + 'B', valor + 'Z', valor+'X']:
        
        # g(n) costo de llegar al nodo
        g = nodo.profundidad+1

        # h(n) Costo de llegar del nodo a la meta (heuristica)
        #h1
        if nuevo_valor[-1]=='B':
            h1 = 1
        else:
            h1 = 2

        #h2
        h2 = abs(len(nuevo_valor)-6) # 6 es el tama√±o de la solucion en este ejemplo

        # Ponderacion
        p = ponderar([(h1, 5), (h2, 10)])

        nuevo_nodo = Nodo(nuevo_valor, padre=nodo, peso=g+p)
        vecinos.append(nuevo_nodo)
    return vecinos
```

```python
nodo_a = Nodo('A', 0)

busqueda_bfs = BFS(nodo_a,goal, F)
busqueda_bfs.search()

print(busqueda_bfs.show_path())


busqueda_dfs = DFS(nodo_a,goal, F, depth=8)
busqueda_dfs.search()

print(busqueda_dfs.show_path())

busqueda_ucs = UCS(nodo_a,goal, F)
busqueda_ucs.search()

print(busqueda_ucs.show_path())
```

```python
print(busqueda_bfs.show_explored())
print(busqueda_dfs.show_explored())
print(busqueda_ucs.show_explored())


```
